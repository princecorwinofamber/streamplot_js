```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

streamplot рисует фазовый портрет, а quiver - просто стрелки на указанных координатах

```python
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
```

```python
plt.streamplot(X, Y, Y, -X)
```

```python
plt.quiver(X, Y, Y, -X)
```

Обрати внимание, даже если передать сетку всего 2x2 точки, streamplot восстанавливает по этой сетке эллипсы, причём на всей области сетки, а quiver рисует всего 4 стрелки в углах - и никаких эллипсов.

```python
x2 = np.linspace(-1, 1, 2)
y2 = np.linspace(-1, 1, 2)
X2, Y2 = np.meshgrid(x2, y2)
```

```python
plt.streamplot(X2, Y2, Y2, -X2)
```

```python
plt.quiver(X2, Y2, Y2, -X2)
```
